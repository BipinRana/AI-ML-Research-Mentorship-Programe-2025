```python
import pandas as pd
import numpy as np
data = pd.read_csv('diabetes.csv')
features = data.iloc[:,:-1]
outcome = data.iloc[:,-1]

```

```python
from sklearn.preprocessing import StandardScaler
features = StandardScaler().fit_transform(features)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(features,outcome, test_size =0.2, random_state = 42)
R_model = LogisticRegression().fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


Y_pred = R_model.predict(X_test)
print(classification_report(Y_test, Y_pred))
```

```python
new_data = np.array([[6,148,72,35,0,33.6,0.627,50]])
print(R_model.predict(new_data))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the plot with both boxplot and stripplot
plt.figure(figsize=(8, 6))

# Boxplot
sns.boxplot(x='Outcome', y='Pregnancies', data=data, color='lightblue')

# Stripplot to show the individual points
sns.stripplot(x='Outcome', y='Pregnancies', data=data, jitter=True, color='black', alpha=0.5)

plt.title('Boxplot and Individual Points of Pregnancies vs Outcome')
plt.xlabel('Outcome (0 = No Diabetes, 1 = Diabetes)')
plt.ylabel('Number of Pregnancies')
plt.show()
```
